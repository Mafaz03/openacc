```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("pade_scheme.dat")
print(data)
```

```python
x    = data[:, 0]
pred = data[:, 1]
actu = data[:, 2]
err  = data[:, 3]

plt.plot(x, pred, label="predicted")
plt.plot(x, actu, label="actual")

plt.grid()
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Padé Scheme Derivative Approximation")
plt.legend()
plt.show()
```
